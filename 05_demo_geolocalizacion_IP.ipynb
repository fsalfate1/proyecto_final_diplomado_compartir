```python
import requests
from geopy.geocoders import Nominatim

# 1) Obtener lat/lon aproximados desde tu IP
ip_data = requests.get("http://ip-api.com/json/").json()
lat, lon = ip_data["lat"], ip_data["lon"]

# 2) Reverse geocoding con Nominatim
geolocator = Nominatim(user_agent="mi_app_direccion")
location = geolocator.reverse((lat, lon), language="es")

print("IP:", ip_data["query"])
print("Ciudad:", ip_data.get("city"))
print("Lat/Lon:", lat, lon)
print("Direccion:", location.address)
```
